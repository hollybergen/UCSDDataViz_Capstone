```python
#function to calculate number of character in str
def count_chars(txt):
    result = 0
    for char in txt:
        result += 1     
    return result
```

```python
#function to determine word count
def count_words(data):
    words = data.split(" ")
    num_words = len(words)
    return num_words
```

```python
test = ['a b','a b c', 'd e']
[count_words(i) for i in test]
```

```python
def string_test(s):
    d={"UPPER_CASE":0, "LOWER_CASE":0}
    uc = d["UPPER_CASE"]
    lc = d["LOWER_CASE"] 

    for c in s:
        if c.isupper():
            uc+=1
        elif c.islower():
            lc+=1
        else:
            pass
    
    return uc, lc
```
