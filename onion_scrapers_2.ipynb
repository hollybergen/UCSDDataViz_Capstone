```python
#function to calculate number of character in str
def count_chars(txt):
    result = 0
    for char in txt:
        result += 1     
    return result
```

```python
#function to determine word count
def count_words(data):
    words = data.split(" ")
    num_words = len(words)
    return num_words
```

```python
def string_test(s):
    d={"UPPER_CASE":0, "LOWER_CASE":0}
    uc = d["UPPER_CASE"]
    lc = d["LOWER_CASE"] 

    for c in s:
        if c.isupper():
            uc+=1
        elif c.islower():
            lc+=1
        else:
            pass
    
    return uc, lc
```

```python
def ucase_test(s):
    uc = 0

    for c in s:
        if c.isupper():
            uc+=1
       
     
#     print ("Original String : ", s)
#     print ("No. of Upper case characters : ", uc)
#     print ("No. of Lower case Characters : ", lc)
    
    return uc
```

```python
def lcase_test(s):
    lc = 0

    for c in s:
        if c.islower():
            lc+=1
       
     
#     print ("Original String : ", s)
#     print ("No. of Upper case characters : ", uc)
#     print ("No. of Lower case Characters : ", lc)
    
    return lc
```

```python
# function checks if the string contains any special character 
def special(str1): 
    spaces = alphabets = digits = special = 0

    for i in range(len(str1)):
        if((str1[i] >= 'a' and str1[i] <= 'z') or (str1[i] >= 'A' and str1[i] <= 'Z')): 
            alphabets = alphabets + 1 
        elif(str1[i] >= '0' and str1[i] <= '9'):
            digits = digits + 1
        elif (str1[i] == ' '):
            spaces = spaces + 1
        else:
            special = special + 1
    return special;
```
