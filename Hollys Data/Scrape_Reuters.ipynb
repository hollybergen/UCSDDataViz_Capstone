```python
from scraper import scrape
```

## Reuters - Top stories

```python
soup = scrape('https://mobile.reuters.com/')
```

```python
print(soup.prettify())
```

```python
top_news = soup.find_all('section', class_="module top-news-module")

articles = top_news[0].find_all('article')
articles[0].find_all('h3')[0].a.text

for article in articles:
    print(article.find_all('h3')[0].a.text)
```

## Reuters - Business

```python
soup = scrape('https://mobile.reuters.com/business')
```

```python
print(soup.prettify())
```

```python
top_news = soup.find_all('section', class_="module top-news-module module-no-margin")
```

```python
articles = top_news[1].find('div', class_= 'module-container').find_all('article')
```

```python
for article in articles:
    print(article.find_all('h3')[0].a.text)
```

```python
bottom_news = soup.find_all('section', class_="module top-news-module module-last")
```

```python
articles = bottom_news[0].find('div', class_= 'module-container').find_all('article')
```

```python
for article in articles:
    print(article.find_all('h3')[0].a.text)
```
